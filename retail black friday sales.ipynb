```python
x = df1.drop(["Purchase"], axis = 1)
y = df1["Purchase"]
```

```python
x.shape
```

```python
x = x.values
y = y.values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
def train(model):
    model.fit(x_train, y_train)
    y_train_pred = model.predict(x_train)
    y_val_pred = model.predict(x_val)
    
    mse = mean_squared_error(y_train, y_train_pred)
    r2 = r2_score(y_train, y_train_pred)
    print('For training data, mean squared error: {:.4f}, R2: {:.4f}'.format(mse, r2))

    mse = mean_squared_error(y_val, y_val_pred)
    r2 = r2_score(y_val, y_val_pred)
    print('For validation data, mean squared error: {:.4f}, R2: {:.4f}'.format(mse, r2))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
train(lr)
```

```python
from sklearn.linear_model import Ridge
rr = Ridge(alpha=0.5)
train(rr)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1)
train(lasso)
```

```python
import statsmodels.formula.api as smf
from sklearn import metrics
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
import warnings 
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

```python
model = Sequential()

# The Input Layer :
model.add(Dense(8, kernel_initializer='normal',input_dim = x_train.shape[1], activation='relu'))

# The Hidden Layers :
model.add(Dense(128, kernel_initializer='normal',activation='relu'))
model.add(Dense(128, kernel_initializer='normal',activation='relu'))
model.add(Dense(128, kernel_initializer='normal',activation='relu'))

# The Output Layer :
model.add(Dense(1, kernel_initializer='normal',activation='linear'))

model.compile(loss='mse', optimizer='adam', metrics=['mse','mae'])
```

```python
history = model.fit(x_train, y_train, epochs=1000, batch_size=256, verbose=1)
```

```python
y_pred = model.predict(x_val)
```

```python
mae = metrics.mean_absolute_error(y_val, y_pred)
mse = metrics.mean_squared_error(y_val, y_pred)
rmse = np.sqrt(metrics.mean_squared_error(y_val, y_pred))
rsquared = metrics.r2_score(y_val, y_pred)
```

```python
model_stat_dict = {
    'Mean Absolute Error': [mae],
    'Mean Squared Error': [mse],
    'Root Mean Squared Error': [rmse],
    'R-squared score': [rsquared]
}
```

```python
model_stat = pd.DataFrame(model_stat_dict)
model_stat
```
